```python
latitude = 37
longitude = -100

# Create map and display it
san_map = folium.Map(location=[latitude, longitude], zoom_start=4)

# Display the map of San Francisco
san_map
```

```python
cdata = pd.read_csv('SAGDP2N__ALL_AREAS_2013_2017.csv')
cdata.head()
```

```python
limit = 200
data = cdata.iloc[0:limit, :]

# Instantiate a feature group for the incidents in the dataframe
incidents = folium.map.FeatureGroup()

# Loop through the 200 crimes and add each to the incidents feature group
for lat, lng, in zip(cdata.Y, data.X):
    incidents.add_child(
        folium.CircleMarker(
            [lat, lng],
            radius=7, # define how big you want the circle markers to be
            color='yellow',
            fill=True,
            fill_color='red',
            fill_opacity=0.4
        )
    )

# Add incidents to map
san_map = folium.Map(location=[latitude, longitude], zoom_start=4)
san_map.add_child(incidents)
```

```python
from folium import plugins

# let's start again with a clean copy of the map of San Francisco
san_map = folium.Map(location = [latitude, longitude], zoom_start = 4)

# instantiate a mark cluster object for the incidents in the dataframe
incidents = plugins.MarkerCluster().add_to(san_map)

# loop through the dataframe and add each data point to the mark cluster
for lat, lng, label, in zip(data.Y, data.X, cdata.GeoName):
    folium.Marker(
        location=[lat, lng],
        icon=None,
        popup=label,
    ).add_to(incidents)

# add incidents to map
san_map.add_child(incidents)
```

```python

import json
import requests

url = 'https://raw.githubusercontent.com/python-visualization/folium/master/examples/data'
san_geo = f'{url}/us-states.json'
san_map = folium.Map(location=[37, -100], zoom_start=4)
folium.GeoJson(
    san_geo,
    style_function=lambda feature: {
        'fillColor': '#ffff00',
        'color': 'black',
        'weight': 2,
        'dashArray': '5, 5'
    }
).add_to(san_map)

#display map
san_map
```

```python
import pandas as pd

disdata = pd.read_csv('SAGDP2N__ALL_AREAS_2013_2017.csv')

san_map = folium.Map(location=[37.77, -122.4], zoom_start=4)

# Create Choropleth map
folium.Choropleth(
    geo_data=san_geo,
    data=disdata,
    columns=['GeoFIPS','Y2017'],
    key_on='feature.id',
    #fill_color='red',
    #fill_color='YlOrRd',
    fill_color='YlGn',
    fill_opacity=0.7,
    line_opacity=0.2,
    highlight=True,
    legend_name='US-States-美国2017年各州GDP'
).add_to(san_map)

san_map
```
